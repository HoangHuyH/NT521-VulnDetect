```python
!pip install torch-geometric
# Transformers is available on Kaggle, update if needed

import json
import torch
import torch.nn as nn
import numpy as np
import os
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel, get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.metrics import (accuracy_score, f1_score, classification_report, 
                             confusion_matrix, roc_auc_score, roc_curve)
from sklearn.model_selection import train_test_split
from tqdm.notebook import tqdm
from torch.cuda.amp import autocast, GradScaler

# --- 1. CONFIGURATION ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"[INFO] Using device: {DEVICE}")

# GPU Optimizations
if torch.cuda.is_available():
    torch.backends.cudnn.benchmark = True
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True
    print(f"[INFO] GPU: {torch.cuda.get_device_name(0)}")
    print(f"[INFO] GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")

# Data paths
PROJECT = "NVD"
INPUT_JSON = f'/kaggle/input/graphfvd-nvd-dataset/NVD/NVD_PrVCs.json'
OUTPUT_DIR = f'/kaggle/working/GraphFVD/dataset/{PROJECT}'

TRAIN_JSONL = os.path.join(OUTPUT_DIR, 'my_train.jsonl')
VALID_JSONL = os.path.join(OUTPUT_DIR, 'my_valid.jsonl')
TEST_JSONL = os.path.join(OUTPUT_DIR, 'my_test.jsonl')

MODEL_NAME = "microsoft/graphcodebert-base"
MAX_LEN = 512 
BATCH_SIZE = 16
EPOCHS = 5
LEARNING_RATE = 2e-5
GRADIENT_ACCUMULATION_STEPS = 2
USE_FP16 = True
NUM_WORKERS = 2

# --- 2. GRAPH FEATURE EXTRACTION ---
def extract_graph_features(item):
    """
    Extract meaningful features from graph structure.
    These features capture graph topology and code patterns that BERT might miss.
    """
    features = []
    
    # === 1. BASIC GRAPH STATISTICS ===
    nodes = item.get('nodes', list(range(len(item.get('nodes_label', [])))))
    edges = item.get('edges', [])
    nodes_label = item.get('nodes_label', [])
    
    num_nodes = len(nodes) if isinstance(nodes, list) else nodes
    if isinstance(nodes_label, list) and len(nodes_label) > 0:
        num_nodes = len(nodes_label)
    num_edges = len(edges)
    
    features.append(num_nodes)
    features.append(num_edges)
    features.append(num_edges / max(num_nodes, 1))  # Edge density
    features.append(num_nodes / max(num_edges, 1))  # Node to edge ratio
    
    # === 2. NODE TYPE DISTRIBUTION ===
    important_types = [
        'FunctionDef', 'IfStatement', 'WhileStatement', 'ForStatement',
        'CallExpression', 'Parameter', 'IdentifierDeclStatement',
        'ReturnStatement', 'SwitchStatement', 'Condition', 'ExpressionStatement',
        'Argument', 'ArgumentList', 'CastExpression', 'Identifier'
    ]
    
    label_counts = {}
    for label in nodes_label:
        label_counts[label] = label_counts.get(label, 0) + 1
    
    for node_type in important_types:
        count = label_counts.get(node_type, 0)
        features.append(count)
        features.append(count / max(num_nodes, 1))  # Normalized
    
    # === 3. DATA FLOW FEATURES ===
    edges_label = item.get('edges_label', [])
    non_empty_labels = [l for l in edges_label if l and str(l).strip()]
    unique_vars = set(non_empty_labels)
    
    features.append(len(unique_vars))  # Number of unique variables
    features.append(len(non_empty_labels))  # Total data flow edges
    features.append(len(non_empty_labels) / max(num_edges, 1))  # Data flow ratio
    
    # Variable frequency statistics
    var_counts = {}
    for var in non_empty_labels:
        var_counts[var] = var_counts.get(var, 0) + 1
    
    if var_counts:
        features.append(max(var_counts.values()))  # Max var usage
        features.append(min(var_counts.values()))  # Min var usage
        features.append(np.mean(list(var_counts.values())))  # Avg var usage
        features.append(np.std(list(var_counts.values())))  # Std var usage
    else:
        features.extend([0, 0, 0, 0])
    
    # === 4. GRAPH STRUCTURE FEATURES ===
    in_degree = {}
    out_degree = {}
    for edge in edges:
        if isinstance(edge, (list, tuple)) and len(edge) >= 2:
            src, dst = edge[0], edge[1]
            out_degree[src] = out_degree.get(src, 0) + 1
            in_degree[dst] = in_degree.get(dst, 0) + 1
    
    # Out-degree statistics
    if out_degree:
        features.append(max(out_degree.values()))
        features.append(np.mean(list(out_degree.values())))
        features.append(np.std(list(out_degree.values())))
        features.append(len([v for v in out_degree.values() if v > 3]))  # High out-degree nodes
    else:
        features.extend([0, 0, 0, 0])
    
    # In-degree statistics
    if in_degree:
        features.append(max(in_degree.values()))
        features.append(np.mean(list(in_degree.values())))
        features.append(np.std(list(in_degree.values())))
        features.append(len([v for v in in_degree.values() if v > 3]))  # High in-degree nodes
    else:
        features.extend([0, 0, 0, 0])
    
    # Leaf nodes (no outgoing edges)
    all_src = set(out_degree.keys())
    all_dst = set(in_degree.keys())
    leaf_nodes = len(all_dst - all_src)
    root_nodes = len(all_src - all_dst)
    features.append(leaf_nodes)
    features.append(root_nodes)
    features.append(leaf_nodes / max(num_nodes, 1))
    
    # === 5. CODE COMPLEXITY FROM nodes_codes ===
    nodes_codes = item.get('nodes_codes', [])
    main_code = nodes_codes[0] if nodes_codes else ""
    
    features.append(len(main_code))  # Code length
    features.append(main_code.count('\n') if main_code else 0)  # Line count
    features.append(main_code.count('if'))  # If statements
    features.append(main_code.count('else'))  # Else statements
    features.append(main_code.count('for'))  # For loops
    features.append(main_code.count('while'))  # While loops
    features.append(main_code.count('switch'))  # Switch statements
    features.append(main_code.count('return'))  # Return statements
    features.append(main_code.count('('))  # Function calls approx
    features.append(main_code.count('{'))  # Block depth indicator
    features.append(main_code.count('*'))  # Pointer usage
    features.append(main_code.count('&'))  # Address operator
    features.append(main_code.count('->'))  # Pointer member access
    features.append(main_code.count('[]') + main_code.count('['))  # Array access
    
    # === 6. VULNERABILITY PATTERNS (Domain Knowledge) ===
    code_lower = main_code.lower() if main_code else ""
    all_codes_str = str(nodes_codes).lower()
    
    # Dangerous functions (common vulnerability sources)
    dangerous_funcs = [
        'strcpy', 'strcat', 'sprintf', 'vsprintf', 'gets', 'scanf',
        'memcpy', 'memmove', 'strncpy', 'strncat', 'atoi', 'atol',
        'realloc', 'alloca', 'system', 'exec', 'popen'
    ]
    for func in dangerous_funcs:
        features.append(1 if func in all_codes_str else 0)
    
    # Memory operations
    features.append(1 if 'malloc' in all_codes_str else 0)
    features.append(1 if 'free' in all_codes_str else 0)
    features.append(1 if 'calloc' in all_codes_str else 0)
    features.append(1 if 'sizeof' in all_codes_str else 0)
    
    # Null checks (good practice indicator)
    features.append(1 if 'null' in all_codes_str else 0)
    features.append(1 if '== null' in all_codes_str or '!= null' in all_codes_str else 0)
    
    # Buffer size indicators
    features.append(1 if 'buffer' in all_codes_str else 0)
    features.append(1 if 'size' in all_codes_str else 0)
    features.append(1 if 'length' in all_codes_str or 'len' in all_codes_str else 0)
    
    return np.array(features, dtype=np.float32)

# Calculate feature dimension dynamically
_sample_features = extract_graph_features({'nodes_label': [], 'edges': [], 'nodes_codes': ['']})
GRAPH_FEATURE_DIM = len(_sample_features)
print(f"[INFO] Graph feature dimension: {GRAPH_FEATURE_DIM}")

# --- 3. PREPROCESS DATA ---
def run_preprocessing():
    """Run preprocessing to create train/valid/test files"""
    print(f"[INFO] Starting preprocessing...")
    
    if not os.path.exists(OUTPUT_DIR):
        os.makedirs(OUTPUT_DIR)
    
    if os.path.exists(TRAIN_JSONL) and os.path.exists(VALID_JSONL):
        print(f"[INFO] Preprocessed files already exist. Skipping.")
        return True
    
    print(f"[INFO] Reading data from: {INPUT_JSON}")
    try:
        js_all = pd.read_json(INPUT_JSON)
        js_all = js_all.to_dict('records')
    except Exception as e:
        print(f"[WARN] Standard read failed: {e}")
        try:
            with open(INPUT_JSON, 'r') as f:
                js_all = [json.loads(line) for line in f]
        except:
            print(f"[ERROR] Cannot read data from {INPUT_JSON}")
            return False
    
    def get_label(d):
        return int(d.get('target', d.get('label_int', d.get('label', d.get('vulnerable', 0)))))
    labels = [get_label(d) for d in js_all]
    
    train_data, temp_data, train_labels, temp_labels = train_test_split(
        js_all, labels, test_size=0.2, stratify=labels, random_state=42
    )
    valid_data, test_data, _, _ = train_test_split(
        temp_data, temp_labels, test_size=0.5, stratify=temp_labels, random_state=42
    )
    
    def save_jsonl(data_list, filename):
        file_path = os.path.join(OUTPUT_DIR, filename)
        with open(file_path, 'w') as f:
            for idx, js in enumerate(data_list):
                js['idx'] = idx
                f.write(json.dumps(js) + '\n')
        print(f"[INFO] Saved to {file_path}")
    
    save_jsonl(train_data, 'my_train.jsonl')
    save_jsonl(valid_data, 'my_valid.jsonl')
    save_jsonl(test_data, 'my_test.jsonl')
    
    print(f"[INFO] Preprocessing done! Train: {len(train_data)}, Valid: {len(valid_data)}, Test: {len(test_data)}")
    return True

# --- 4. DATASET WITH GRAPH FEATURES ---
class GraphCodeDataset(Dataset):
    """Dataset that combines code text with graph features"""
    
    def __init__(self, data, tokenizer, max_len):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len
        
        # Pre-compute graph features for all samples
        print("[INFO] Extracting graph features...")
        self.graph_features = []
        for item in tqdm(data, desc="Graph Features"):
            features = extract_graph_features(item)
            self.graph_features.append(features)
        
        # Normalize features
        self.graph_features = np.array(self.graph_features)
        self.feature_mean = np.mean(self.graph_features, axis=0)
        self.feature_std = np.std(self.graph_features, axis=0) + 1e-8
        self.graph_features = (self.graph_features - self.feature_mean) / self.feature_std
        
        print(f"[INFO] Graph features shape: {self.graph_features.shape}")

    def __len__(self):
        return len(self.data)
    
    def extract_code_from_graph(self, item):
        if 'nodes_codes' in item and len(item['nodes_codes']) > 0:
            return item['nodes_codes'][0]
        return str(item.get('code', item.get('func', item.get('function', ''))))
    
    def extract_dataflow_from_graph(self, item):
        if 'edges_label' in item:
            edges_label = item['edges_label']
            var_names = [label for label in edges_label if label and str(label).strip()]
            var_names = list(set(var_names))[:64]
            return " ".join(var_names)
        return ""

    def __getitem__(self, idx):
        item = self.data[idx]
        
        code = self.extract_code_from_graph(item)
        data_flow_str = self.extract_dataflow_from_graph(item)
        
        label = int(item.get('target', item.get('label_int', item.get('label', item.get('vulnerable', 0)))))
        
        encoding = self.tokenizer(
            code,
            data_flow_str, 
            truncation=True,
            max_length=self.max_len,
            padding='max_length',
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'graph_features': torch.tensor(self.graph_features[idx], dtype=torch.float32),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# --- 5. GRAPH-ENHANCED BERT MODEL ---
class GraphEnhancedBERT(nn.Module):
    """
    Hybrid model combining GraphCodeBERT with handcrafted graph features.
    - GraphCodeBERT: Captures code semantics and data flow from text
    - Graph Features: Captures graph topology and vulnerability patterns
    """
    
    def __init__(self, model_name, graph_feature_dim, num_labels=2, dropout=0.3):
        super().__init__()
        
        # GraphCodeBERT backbone
        self.bert = AutoModel.from_pretrained(model_name)
        bert_hidden = self.bert.config.hidden_size  # 768
        
        # Freeze some BERT layers for faster training (optional)
        # for param in self.bert.embeddings.parameters():
        #     param.requires_grad = False
        
        # Graph feature encoder
        self.graph_encoder = nn.Sequential(
            nn.Linear(graph_feature_dim, 256),
            nn.LayerNorm(256),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(256, 128),
            nn.LayerNorm(128),
            nn.GELU(),
        )
        
        # Attention-based fusion
        self.fusion_attention = nn.Sequential(
            nn.Linear(bert_hidden + 128, 256),
            nn.Tanh(),
            nn.Linear(256, 2),
            nn.Softmax(dim=-1)
        )
        
        # Feature fusion layer
        self.fusion = nn.Sequential(
            nn.Linear(bert_hidden + 128, 512),
            nn.LayerNorm(512),
            nn.GELU(),
            nn.Dropout(dropout),
        )
        
        # Classification head
        self.classifier = nn.Sequential(
            nn.Linear(512, 256),
            nn.LayerNorm(256),
            nn.GELU(),
            nn.Dropout(dropout * 0.7),
            nn.Linear(256, 64),
            nn.GELU(),
            nn.Linear(64, num_labels)
        )
        
        # Initialize weights
        self._init_weights()
    
    def _init_weights(self):
        for module in [self.graph_encoder, self.fusion, self.classifier]:
            for m in module.modules():
                if isinstance(m, nn.Linear):
                    nn.init.xavier_uniform_(m.weight)
                    if m.bias is not None:
                        nn.init.zeros_(m.bias)
    
    def forward(self, input_ids, attention_mask, graph_features):
        # BERT encoding
        bert_output = self.bert(input_ids, attention_mask=attention_mask)
        cls_embedding = bert_output.last_hidden_state[:, 0, :]  # [CLS] token: (B, 768)
        
        # Graph feature encoding
        graph_embedding = self.graph_encoder(graph_features)  # (B, 128)
        
        # Concatenate features
        combined = torch.cat([cls_embedding, graph_embedding], dim=-1)  # (B, 896)
        
        # Fusion
        fused = self.fusion(combined)  # (B, 512)
        
        # Classification
        logits = self.classifier(fused)  # (B, 2)
        
        return logits

# --- 6. LOAD DATA ---
def load_jsonl_file(file_path):
    data_list = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            try:
                data_list.append(json.loads(line))
            except:
                pass
    return data_list

def load_and_prepare_data():
    if not run_preprocessing():
        print("[ERROR] Preprocessing failed!")
        return [], []
    
    print(f"[INFO] Loading training data...")
    train_data = load_jsonl_file(TRAIN_JSONL)
    
    print(f"[INFO] Loading validation data...")
    val_data = load_jsonl_file(VALID_JSONL)
    
    print(f"[INFO] Data loaded. Train: {len(train_data)}, Val: {len(val_data)}")
    return train_data, val_data

# --- 7. VISUALIZATION ---
def plot_confusion_matrix(y_true, y_pred, epoch, save_path=None):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Benign', 'Vulnerable'],
                yticklabels=['Benign', 'Vulnerable'])
    plt.title(f'Confusion Matrix - Epoch {epoch}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()
    return cm

def plot_roc_curve(y_true, y_probs, epoch, save_path=None):
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    auc = roc_auc_score(y_true, y_probs)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - Epoch {epoch}')
    plt.legend(loc='lower right')
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()
    return auc

def plot_training_history(history, save_path=None):
    epochs = range(1, len(history['accuracy']) + 1)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    axes[0, 0].plot(epochs, history['accuracy'], 'b-o', label='Accuracy')
    axes[0, 0].set_title('Validation Accuracy')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Accuracy')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    
    axes[0, 1].plot(epochs, history['f1'], 'g-o', label='F1 Score')
    axes[0, 1].set_title('Validation F1 Score')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('F1 Score')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    
    axes[1, 0].plot(epochs, history['auc'], 'r-o', label='AUC')
    axes[1, 0].set_title('Validation AUC')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('AUC')
    axes[1, 0].legend()
    axes[1, 0].grid(True)
    
    axes[1, 1].plot(epochs, history['train_loss'], 'm-o', label='Train Loss')
    axes[1, 1].set_title('Training Loss')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Loss')
    axes[1, 1].legend()
    axes[1, 1].grid(True)
    
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()

# --- 8. TRAINING LOOP ---
def run_training():
    train_data, val_data = load_and_prepare_data()
    if not train_data:
        return None, None

    print(f"[INFO] Loading tokenizer: {MODEL_NAME}...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
    
    print(f"[INFO] Creating model: GraphEnhancedBERT...")
    model = GraphEnhancedBERT(MODEL_NAME, GRAPH_FEATURE_DIM, num_labels=2)
    model.to(DEVICE)
    
    # Count parameters
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"[INFO] Total parameters: {total_params:,}")
    print(f"[INFO] Trainable parameters: {trainable_params:,}")

    # Create datasets
    train_dataset = GraphCodeDataset(train_data, tokenizer, MAX_LEN)
    val_dataset = GraphCodeDataset(val_data, tokenizer, MAX_LEN)
    
    # Store normalization params for inference
    feature_mean = train_dataset.feature_mean
    feature_std = train_dataset.feature_std

    # DataLoaders
    train_loader = DataLoader(
        train_dataset, 
        batch_size=BATCH_SIZE, 
        shuffle=True,
        num_workers=NUM_WORKERS,
        pin_memory=True if torch.cuda.is_available() else False
    )
    val_loader = DataLoader(
        val_dataset, 
        batch_size=BATCH_SIZE * 2,
        num_workers=NUM_WORKERS,
        pin_memory=True if torch.cuda.is_available() else False
    )

    # Optimizer
    optimizer = AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=0.01)
    total_steps = (len(train_loader) // GRADIENT_ACCUMULATION_STEPS) * EPOCHS
    scheduler = get_linear_schedule_with_warmup(
        optimizer, 
        num_warmup_steps=int(0.1 * total_steps), 
        num_training_steps=total_steps
    )
    
    # Mixed Precision
    scaler = GradScaler() if USE_FP16 and torch.cuda.is_available() else None
    if scaler:
        print("[INFO] Using Mixed Precision Training (FP16)")
    
    # Class Weights
    def get_label(d):
        return int(d.get('target', d.get('label_int', d.get('label', d.get('vulnerable', 0)))))
    y_train = [get_label(d) for d in train_data]
    n0 = y_train.count(0)
    n1 = y_train.count(1)
    weight_val = n0 / n1 if n1 > 0 else 1.0
    class_weights = torch.tensor([1.0, weight_val]).to(DEVICE)
    print(f"[INFO] Class distribution: Benign={n0}, Vulnerable={n1}")
    print(f"[INFO] Class Weights: {class_weights}")
    
    loss_fn = nn.CrossEntropyLoss(weight=class_weights)

    best_f1 = 0.0
    history = {'accuracy': [], 'f1': [], 'auc': [], 'train_loss': []}
    
    print("\n" + "="*60)
    print("START TRAINING GRAPH-ENHANCED BERT")
    print("="*60)
    
    for epoch in range(EPOCHS):
        print(f"\n--- Epoch {epoch+1}/{EPOCHS} ---")
        
        # === TRAIN ===
        model.train()
        train_loss = 0
        optimizer.zero_grad()
        train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Train]")
        
        for step, batch in enumerate(train_bar):
            input_ids = batch['input_ids'].to(DEVICE, non_blocking=True)
            attention_mask = batch['attention_mask'].to(DEVICE, non_blocking=True)
            graph_features = batch['graph_features'].to(DEVICE, non_blocking=True)
            labels = batch['labels'].to(DEVICE, non_blocking=True)

            if scaler:
                with autocast():
                    logits = model(input_ids, attention_mask, graph_features)
                    loss = loss_fn(logits, labels)
                    loss = loss / GRADIENT_ACCUMULATION_STEPS
                scaler.scale(loss).backward()
            else:
                logits = model(input_ids, attention_mask, graph_features)
                loss = loss_fn(logits, labels)
                loss = loss / GRADIENT_ACCUMULATION_STEPS
                loss.backward()
            
            train_loss += loss.item() * GRADIENT_ACCUMULATION_STEPS
            
            if (step + 1) % GRADIENT_ACCUMULATION_STEPS == 0:
                if scaler:
                    scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    optimizer.step()
                
                scheduler.step()
                optimizer.zero_grad()
            
            train_bar.set_postfix({'loss': f"{loss.item() * GRADIENT_ACCUMULATION_STEPS:.4f}"})
        
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        
        avg_train_loss = train_loss / len(train_loader)
        history['train_loss'].append(avg_train_loss)

        # === VALIDATE ===
        model.eval()
        val_preds = []
        val_probs = []
        val_labels_list = []
        
        print("   Validating...")
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(DEVICE, non_blocking=True)
                attention_mask = batch['attention_mask'].to(DEVICE, non_blocking=True)
                graph_features = batch['graph_features'].to(DEVICE, non_blocking=True)
                labels = batch['labels'].to(DEVICE, non_blocking=True)
                
                if scaler:
                    with autocast():
                        logits = model(input_ids, attention_mask, graph_features)
                else:
                    logits = model(input_ids, attention_mask, graph_features)
                
                probs = torch.softmax(logits, dim=1)
                preds = torch.argmax(logits, dim=1)
                
                val_preds.extend(preds.cpu().numpy())
                val_probs.extend(probs[:, 1].cpu().numpy())
                val_labels_list.extend(labels.cpu().numpy())

        # Calculate metrics
        acc = accuracy_score(val_labels_list, val_preds)
        f1 = f1_score(val_labels_list, val_preds)
        
        try:
            auc = roc_auc_score(val_labels_list, val_probs)
        except ValueError:
            auc = 0.5
        
        history['accuracy'].append(acc)
        history['f1'].append(f1)
        history['auc'].append(auc)
        
        # Print results
        print(f"\n{'='*50}")
        print(f"EPOCH {epoch+1} RESULTS")
        print(f"{'='*50}")
        print(f"Train Loss:     {avg_train_loss:.4f}")
        print(f"Val Accuracy:   {acc:.4f}")
        print(f"Val F1 Score:   {f1:.4f}")
        print(f"Val AUC:        {auc:.4f}")
        print(f"{'='*50}")
        
        print("\nClassification Report:")
        print(classification_report(val_labels_list, val_preds, target_names=['Benign', 'Vulnerable']))
        
        # Plot confusion matrix
        plot_confusion_matrix(val_labels_list, val_preds, epoch+1, 
                             save_path=f"/kaggle/working/confusion_matrix_epoch_{epoch+1}.png")
        
        # Plot ROC curve
        plot_roc_curve(val_labels_list, val_probs, epoch+1,
                      save_path=f"/kaggle/working/roc_curve_epoch_{epoch+1}.png")

        if f1 > best_f1:
            best_f1 = f1
            save_dict = {
                'model_state_dict': model.state_dict(),
                'feature_mean': feature_mean,
                'feature_std': feature_std,
                'graph_feature_dim': GRAPH_FEATURE_DIM,
            }
            torch.save(save_dict, "/kaggle/working/best_graph_enhanced_bert.pth")
            print("[INFO] New Best Model Saved!")
    
    # Plot training history
    print("\n" + "="*60)
    print("TRAINING HISTORY")
    print("="*60)
    plot_training_history(history, save_path="/kaggle/working/training_history.png")
    
    print(f"\n[INFO] Training completed! Best F1: {best_f1:.4f}")
    return model, history

# Run training
model, history = run_training()
```
