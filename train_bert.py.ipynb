```python
!pip install torch-geometric
# Transformers is available on Kaggle, update if needed

import json
import torch
import numpy as np
import os
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.metrics import (accuracy_score, f1_score, classification_report, 
                             confusion_matrix, roc_auc_score, roc_curve)
from sklearn.model_selection import train_test_split
from tqdm.notebook import tqdm
from torch.cuda.amp import autocast, GradScaler  # Mixed Precision

# --- 1. CONFIGURATION ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"[INFO] Using device: {DEVICE}")

# GPU Optimizations
if torch.cuda.is_available():
    torch.backends.cudnn.benchmark = True  # Auto-tune for GPU
    torch.backends.cuda.matmul.allow_tf32 = True  # TensorFloat-32 for Ampere GPUs
    torch.backends.cudnn.allow_tf32 = True
    print(f"[INFO] GPU: {torch.cuda.get_device_name(0)}")
    print(f"[INFO] GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")

# Data paths from my_preprocess.py
PROJECT = "NVD"
INPUT_JSON = f'/kaggle/input/graphfvd-nvd-dataset/NVD/NVD_PrVCs.json'
OUTPUT_DIR = f'/kaggle/working/GraphFVD/dataset/{PROJECT}'

# Preprocessed file paths
TRAIN_JSONL = os.path.join(OUTPUT_DIR, 'my_train.jsonl')
VALID_JSONL = os.path.join(OUTPUT_DIR, 'my_valid.jsonl')
TEST_JSONL = os.path.join(OUTPUT_DIR, 'my_test.jsonl')

MODEL_NAME = "microsoft/graphcodebert-base"
MAX_LEN = 512 
BATCH_SIZE = 16  # Increased for GPU
EPOCHS = 3
LEARNING_RATE = 2e-5
GRADIENT_ACCUMULATION_STEPS = 2  # Effective batch size = BATCH_SIZE * GRADIENT_ACCUMULATION_STEPS
USE_FP16 = True  # Mixed Precision Training
NUM_WORKERS = 2  # DataLoader workers

# --- 2. PREPROCESS DATA ---
def run_preprocessing():
    """Run preprocessing to create train/valid/test files"""
    print(f"[INFO] Starting preprocessing...")
    
    # Create output directory
    if not os.path.exists(OUTPUT_DIR):
        os.makedirs(OUTPUT_DIR)
    
    # Check if files already exist
    if os.path.exists(TRAIN_JSONL) and os.path.exists(VALID_JSONL):
        print(f"[INFO] Preprocessed files already exist. Skipping this step.")
        return True
    
    # Read data
    print(f"[INFO] Reading data from: {INPUT_JSON}")
    try:
        # Try reading as dataframe
        js_all = pd.read_json(INPUT_JSON)
        js_all = js_all.to_dict('records')
    except Exception as e:
        print(f"[WARN] Standard read failed, trying line-by-line: {e}")
        # If file is JSONL format
        try:
            with open(INPUT_JSON, 'r') as f:
                js_all = [json.loads(line) for line in f]
        except:
            print(f"[ERROR] Cannot read data from {INPUT_JSON}")
            return False
    
    # Split data Train/Valid/Test using train_test_split
    # Get labels for stratify
    def get_label(d):
        # Priority 'target' as it's the key in graph data
        return int(d.get('target', d.get('label_int', d.get('label', d.get('vulnerable', 0)))))
    labels = [get_label(d) for d in js_all]
    
    # Split: 80% train, 10% valid, 10% test
    train_data, temp_data, train_labels, temp_labels = train_test_split(
        js_all, labels, test_size=0.2, stratify=labels, random_state=42
    )
    valid_data, test_data, valid_labels, test_labels = train_test_split(
        temp_data, temp_labels, test_size=0.5, stratify=temp_labels, random_state=42
    )
    
    # Save function
    def save_jsonl(data_list, filename):
        file_path = os.path.join(OUTPUT_DIR, filename)
        with open(file_path, 'w') as f:
            for idx, js in enumerate(data_list):
                js['idx'] = idx
                f.write(json.dumps(js) + '\n')
        print(f"[INFO] Saved to {file_path}")
    
    save_jsonl(train_data, 'my_train.jsonl')
    save_jsonl(valid_data, 'my_valid.jsonl')
    save_jsonl(test_data, 'my_test.jsonl')
    
    print(f"[INFO] Preprocessing completed! Total: {len(js_all)}, Train: {len(train_data)}, Valid: {len(valid_data)}, Test: {len(test_data)}")
    return True

# --- 3. DATASET FOR GRAPH DATA ---
class GraphCodeDataset(Dataset):
    """
    Dataset to process graph data with structure:
    - nodes_codes: Code for each node in graph
    - edges: Edges connecting [source, target]
    - edges_label: Variable names representing data flow
    - target: Label (0=benign, 1=vulnerable)
    """
    def __init__(self, data, tokenizer, max_len):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)
    
    def extract_code_from_graph(self, item):
        """Extract code from graph nodes"""
        # If nodes_codes exists, get code from there
        if 'nodes_codes' in item:
            nodes_codes = item['nodes_codes']
            # First node is usually FunctionDef containing the entire function
            if len(nodes_codes) > 0:
                # Get code of function definition (node 0)
                code = nodes_codes[0]
                return code
        
        # Fallback: use common keys
        return str(item.get('code', item.get('func', item.get('function', ''))))
    
    def extract_dataflow_from_graph(self, item):
        """Extract data flow from edges_label"""
        if 'edges_label' in item:
            edges_label = item['edges_label']
            # Filter non-empty labels (variable names in data flow)
            var_names = [label for label in edges_label if label and label.strip()]
            # Get unique and limit count
            var_names = list(set(var_names))[:64]
            return " ".join(var_names)
        
        # Fallback: use old DFG if available
        dfg_nodes = item.get('dfg', {}).get('nodes', {})
        if dfg_nodes:
            var_names = [str(info.get('name', '')) for info in dfg_nodes.values()]
            var_names = list(set(var_names))[:64]
            return " ".join(var_names)
        
        return ""
    
    def extract_graph_structure(self, item):
        """Extract graph structure info to enhance context"""
        extra_info = []
        
        # Get node labels (node types in AST)
        if 'nodes_label' in item:
            node_types = list(set(item['nodes_label']))[:32]
            extra_info.extend(node_types)
        
        # Get important statements from nodes_codes
        if 'nodes_codes' in item:
            # Get important code snippets (not function def)
            important_codes = []
            for i, code in enumerate(item['nodes_codes'][1:], 1):  # Skip node 0 (already used as main code)
                if code and len(code) < 100:  # Only get short code
                    important_codes.append(code)
            extra_info.extend(important_codes[:20])  # Limit to 20 snippets
        
        return " [SEP] ".join(extra_info) if extra_info else ""

    def __getitem__(self, idx):
        item = self.data[idx]
        
        # 1. Extract code from graph
        code = self.extract_code_from_graph(item)
        
        # 2. Extract data flow from edges
        data_flow_str = self.extract_dataflow_from_graph(item)
        
        # 3. Extract additional context from graph structure
        graph_context = self.extract_graph_structure(item)
        
        # Combine data flow and graph context
        if graph_context:
            data_flow_str = f"{data_flow_str} [SEP] {graph_context}"
        
        # 4. Get label
        label = int(item.get('target', item.get('label_int', item.get('label', item.get('vulnerable', 0)))))
        
        # 5. Tokenize for GraphCodeBERT
        encoding = self.tokenizer(
            code,
            data_flow_str, 
            truncation=True,
            max_length=self.max_len,
            padding='max_length',
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# --- 4. LOAD DATA FROM PREPROCESSED FILES ---
def load_jsonl_file(file_path):
    """Read data from JSONL file"""
    data_list = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            try:
                data_list.append(json.loads(line))
            except:
                pass
    return data_list

def load_and_prepare_data():
    """Load data from preprocessed files by my_preprocess.py"""
    
    # Run preprocessing if needed
    if not run_preprocessing():
        print("[ERROR] Preprocessing failed!")
        return [], []
    
    # Read train and valid data from preprocessed files
    print(f"[INFO] Loading training data from {TRAIN_JSONL}...")
    train_data = load_jsonl_file(TRAIN_JSONL)
    
    print(f"[INFO] Loading validation data from {VALID_JSONL}...")
    val_data = load_jsonl_file(VALID_JSONL)
    
    print(f"[INFO] Data loaded. Train: {len(train_data)}, Val: {len(val_data)}")
    return train_data, val_data

# --- 5. VISUALIZATION FUNCTIONS ---
def plot_confusion_matrix(y_true, y_pred, epoch, save_path=None):
    """Plot confusion matrix"""
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Benign', 'Vulnerable'],
                yticklabels=['Benign', 'Vulnerable'])
    plt.title(f'Confusion Matrix - Epoch {epoch}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()
    return cm

def plot_roc_curve(y_true, y_probs, epoch, save_path=None):
    """Plot ROC curve"""
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    auc = roc_auc_score(y_true, y_probs)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - Epoch {epoch}')
    plt.legend(loc='lower right')
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()
    return auc

def plot_training_history(history, save_path=None):
    """Plot training history with all metrics"""
    epochs = range(1, len(history['accuracy']) + 1)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Accuracy
    axes[0, 0].plot(epochs, history['accuracy'], 'b-o', label='Accuracy')
    axes[0, 0].set_title('Validation Accuracy')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Accuracy')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    
    # F1 Score
    axes[0, 1].plot(epochs, history['f1'], 'g-o', label='F1 Score')
    axes[0, 1].set_title('Validation F1 Score')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('F1 Score')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    
    # AUC
    axes[1, 0].plot(epochs, history['auc'], 'r-o', label='AUC')
    axes[1, 0].set_title('Validation AUC')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('AUC')
    axes[1, 0].legend()
    axes[1, 0].grid(True)
    
    # Training Loss
    axes[1, 1].plot(epochs, history['train_loss'], 'm-o', label='Train Loss')
    axes[1, 1].set_title('Training Loss')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Loss')
    axes[1, 1].legend()
    axes[1, 1].grid(True)
    
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()

# --- 6. TRAINING LOOP ---
def run_graphcodebert_finetuning():
    train_data, val_data = load_and_prepare_data()
    if not train_data: return

    print(f"[INFO] Loading Model: {MODEL_NAME}...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
    model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)
    model.to(DEVICE)

    train_dataset = GraphCodeDataset(train_data, tokenizer, MAX_LEN)
    val_dataset = GraphCodeDataset(val_data, tokenizer, MAX_LEN)

    # Optimized DataLoader for GPU
    train_loader = DataLoader(
        train_dataset, 
        batch_size=BATCH_SIZE, 
        shuffle=True,
        num_workers=NUM_WORKERS,
        pin_memory=True if torch.cuda.is_available() else False
    )
    val_loader = DataLoader(
        val_dataset, 
        batch_size=BATCH_SIZE * 2,  # Larger batch for validation (no gradients)
        num_workers=NUM_WORKERS,
        pin_memory=True if torch.cuda.is_available() else False
    )

    optimizer = AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=0.01)
    total_steps = (len(train_loader) // GRADIENT_ACCUMULATION_STEPS) * EPOCHS
    scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=int(0.1*total_steps), num_training_steps=total_steps)
    
    # Mixed Precision Scaler
    scaler = GradScaler() if USE_FP16 and torch.cuda.is_available() else None
    if scaler:
        print("[INFO] Using Mixed Precision Training (FP16)")
    
    # Class Weights - support multiple label field names
    def get_label(d):
        # Priority 'target' as it's the key in graph data
        return int(d.get('target', d.get('label_int', d.get('label', d.get('vulnerable', 0)))))
    y_train = [get_label(d) for d in train_data]
    n0 = y_train.count(0)
    n1 = y_train.count(1)
    weight_val = n0 / n1 if n1 > 0 else 1.0
    class_weights = torch.tensor([1.0, weight_val]).to(DEVICE)
    print(f"[INFO] Class Weights: {class_weights}")
    
    loss_fn = torch.nn.CrossEntropyLoss(weight=class_weights)

    best_f1 = 0.0
    
    # History for plotting
    history = {
        'accuracy': [],
        'f1': [],
        'auc': [],
        'train_loss': []
    }
    
    print("\n" + "="*60)
    print("START FINE-TUNING GRAPHCODEBERT")
    print("="*60)
    
    for epoch in range(EPOCHS):
        print(f"\n--- Epoch {epoch+1}/{EPOCHS} ---")
        
        # --- TRAIN ---
        model.train()
        train_loss = 0
        optimizer.zero_grad()  # Move outside loop for gradient accumulation
        train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Train]")
        
        for step, batch in enumerate(train_bar):
            input_ids = batch['input_ids'].to(DEVICE, non_blocking=True)
            attention_mask = batch['attention_mask'].to(DEVICE, non_blocking=True)
            labels = batch['labels'].to(DEVICE, non_blocking=True)

            # Mixed Precision Forward Pass
            if scaler:
                with autocast():
                    outputs = model(input_ids, attention_mask=attention_mask)
                    loss = loss_fn(outputs.logits, labels)
                    loss = loss / GRADIENT_ACCUMULATION_STEPS  # Scale loss for accumulation
                
                scaler.scale(loss).backward()
            else:
                outputs = model(input_ids, attention_mask=attention_mask)
                loss = loss_fn(outputs.logits, labels)
                loss = loss / GRADIENT_ACCUMULATION_STEPS
                loss.backward()
            
            train_loss += loss.item() * GRADIENT_ACCUMULATION_STEPS
            
            # Gradient Accumulation: Update weights every N steps
            if (step + 1) % GRADIENT_ACCUMULATION_STEPS == 0:
                if scaler:
                    scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    optimizer.step()
                
                scheduler.step()
                optimizer.zero_grad()
            
            train_bar.set_postfix({'loss': f"{loss.item() * GRADIENT_ACCUMULATION_STEPS:.4f}"})
        
        # Clear GPU cache after training
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        
        avg_train_loss = train_loss / len(train_loader)
        history['train_loss'].append(avg_train_loss)

        # --- VALIDATE ---
        model.eval()
        val_preds = []
        val_probs = []
        val_labels_list = []
        
        print("   Validating...")
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(DEVICE, non_blocking=True)
                attention_mask = batch['attention_mask'].to(DEVICE, non_blocking=True)
                labels = batch['labels'].to(DEVICE, non_blocking=True)
                
                # Mixed Precision Inference
                if scaler:
                    with autocast():
                        outputs = model(input_ids, attention_mask=attention_mask)
                else:
                    outputs = model(input_ids, attention_mask=attention_mask)
                
                probs = torch.softmax(outputs.logits, dim=1)
                preds = torch.argmax(outputs.logits, dim=1)
                
                val_preds.extend(preds.cpu().numpy())
                val_probs.extend(probs[:, 1].cpu().numpy())  # Probability of class 1 (Vulnerable)
                val_labels_list.extend(labels.cpu().numpy())

        # Calculate metrics
        acc = accuracy_score(val_labels_list, val_preds)
        f1 = f1_score(val_labels_list, val_preds)
        
        # Calculate AUC (handle edge case when only one class present)
        try:
            auc = roc_auc_score(val_labels_list, val_probs)
        except ValueError:
            auc = 0.5  # Default when only one class is present
        
        # Save to history
        history['accuracy'].append(acc)
        history['f1'].append(f1)
        history['auc'].append(auc)
        
        # Print results
        print(f"\n{'='*50}")
        print(f"EPOCH {epoch+1} RESULTS")
        print(f"{'='*50}")
        print(f"Train Loss:     {avg_train_loss:.4f}")
        print(f"Val Accuracy:   {acc:.4f}")
        print(f"Val F1 Score:   {f1:.4f}")
        print(f"Val AUC:        {auc:.4f}")
        print(f"{'='*50}")
        
        # Print classification report
        print("\nClassification Report:")
        print(classification_report(val_labels_list, val_preds, target_names=['Benign', 'Vulnerable']))
        
        # Plot confusion matrix
        print("\nConfusion Matrix:")
        cm = plot_confusion_matrix(val_labels_list, val_preds, epoch+1, 
                                   save_path=f"/kaggle/working/confusion_matrix_epoch_{epoch+1}.png")
        
        # Plot ROC curve
        print("\nROC Curve:")
        plot_roc_curve(val_labels_list, val_probs, epoch+1,
                       save_path=f"/kaggle/working/roc_curve_epoch_{epoch+1}.png")

        if f1 > best_f1:
            best_f1 = f1
            torch.save(model.state_dict(), "/kaggle/working/best_graphcodebert_finetuned.pth")
            print("[INFO] New Best Model Saved!")
    
    # Plot training history
    print("\n" + "="*60)
    print("TRAINING HISTORY")
    print("="*60)
    plot_training_history(history, save_path="/kaggle/working/training_history.png")
    
    print(f"\n[INFO] Training completed! Best F1: {best_f1:.4f}")
    return model, history

run_graphcodebert_finetuning()
```
